```python
# Given dictionary
data = statsapi.player_stat_data(statsapi.lookup_player('ohtani, shohei')[0]['id'], group="hitting", type="yearByYear", sportId=1)

# Flatten the dictionary and extract stats for each season
player_stats = []
for season_stats in data['stats']:
    season_data = season_stats['stats']
    season_data['season'] = season_stats['season']
    player_stats.append(season_data)

# Create DataFrame from the list of dictionaries
ohtani_df = pd.DataFrame(player_stats)

# Display DataFrame
ohtani_df
```

```python
#Examine Ohtani pitching stats next

pitch_dict = {'Ohtani': statsapi.lookup_player('ohtani, shohei')[0]['id'], 'Cole':statsapi.lookup_player('cole, gerrit')[0]['id']}

data = []
for player_name, player_id in pitch_dict.items():  # Iterate over player names and IDs
    player_data = statsapi.player_stat_data(player_id, group="pitching", type="career", sportId=1)
    player_data['player_name'] = player_name  # Add player name to the data dictionary
    data.append(player_data)

# Flatten the dictionary and extract stats for each season
player_stats = []
for d in data:
    for season_stats in d['stats']:
        season_data = season_stats['stats']
        season_data['season'] = season_stats['season']
        season_data['player_name'] = d['player_name']  # Add player name to each season's data
        player_stats.append(season_data)

# Create DataFrame from the list of dictionaries
ohtani_pitch = pd.DataFrame(player_stats)

# Display DataFrame
ohtani_pitch
```

```python
column_name = 'player_name'

# Pop the column from its current position
column = ohtani_pitch.pop(column_name)

# Insert the column at the beginning of the DataFrame
ohtani_pitch.insert(0, column_name, column)

columns_to_keep = ['player_name', 'gamesStarted', 'era', 'avg', 'ops', 'whip', 'strikeoutWalkRatio', 'walksPer9Inn', 'hitsPer9Inn', 'runsScoredPer9']

ohtani_pitch = ohtani_pitch.loc[:, columns_to_keep]
ohtani_pitch
```
