```python
import statsapi
import pandas as pd
from pandas import Series, DataFrame
```

```python
# Given dictionary
data = statsapi.player_stat_data(statsapi.lookup_player('ohtani, shohei')[0]['id'], group="pitching", type="yearByYear", sportId=1)

# Flatten the dictionary and extract stats for each season
player_stats = []
for season_stats in data['stats']:
    season_data = season_stats['stats']
    season_data['season'] = season_stats['season']
    player_stats.append(season_data)

# Create DataFrame from the list of dictionaries
ohtani_df = pd.DataFrame(player_stats)

# Display DataFrame
ohtani_df
```

```python
# Column to move to the front
column_to_move = 'season'

# Extract the column
column = ohtani_df[column_to_move]

# Drop the column from its original position
ohtani_df.drop(columns=[column_to_move], inplace=True)

# Insert the column at the front
ohtani_df.insert(0, column_to_move, column)

ohtani_df.insert(0, 'name', 'Shohei Ohtani')

# Display the DataFrame
ohtani_df
```

```python
columns_to_keep = ['name', 'season', 'gamesStarted', 'wins', 'losses', 'inningsPitched', 'era', 'earnedRuns', 'avg', 'ops', 'whip', 'strikeOuts', 'strikeoutWalkRatio', 'totalBases', 'walksPer9Inn', 'hitsPer9Inn', 'runsScoredPer9']
ohtani_pitch = ohtani_df.loc[:, columns_to_keep]
ohtani_pitch
```

```python
ohtani_pitch_comp = ohtani_pitch[ohtani_pitch['season'] == '2022']
ohtani_pitch_comp
```
